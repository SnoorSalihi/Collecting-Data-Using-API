## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
#import all libraries
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

# use the inline backend to generate the plots within the browser
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')  # optional: for ggplot-like style
```

```python
#lets have a quick look at Data
QUERY = """
SELECT ConvertedComp
FROM master
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head(2)
```

Next, we use Matplotlib to create a histogram of the ConvertedComp column. We specify the number of bins to use in the histogram (30, in this case), and set the title, x-axis label, and y-axis label

```python
# Create a histogram of the ConvertedComp column
plt.hist(df['ConvertedComp'], bins=30)
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
#lets have a quick look at Data
QUERY = """
SELECT Age
FROM master
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
#first we will check if there is any null value in the 'Age' column
print(df.isnull().sum()) # check for missing values
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
#lets have a quick look at Data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head(2)
```

A scatter plot (2D) is a useful method of comparing variables against each other.

```python
# your code goes here

df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('Total Work week hours')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.ylim(0, 300)  # replace 0 and 100 with your desired range
plt.show()
```

```python
# calculate the correlation coefficient
corr = df["Age"].corr(df["WorkWeekHrs"])

# print the correlation coefficient
print("Correlation coefficient:", corr)
```

```python
sns.regplot(x='Age', y='WorkWeekHrs', data=df)
plt.ylim(0,)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
#lets have a quick look at Data
QUERY = """
SELECT DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# Count the number of occurrences of each database and select the top 5
top_databases = df['DatabaseDesireNextYear'].value_counts().nlargest(5)

# Create a pie chart with the top 5 databases
fig, ax = plt.subplots()
ax.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%')
ax.set_title('Top 5 databases respondents wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
#lets have a quick look at Data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master where Age >=30 AND Age <=35 Group By CodeRevHrs
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# Create a stacked bar chart of the pivot table
df.plot(kind='bar', stacked=True)

# Set the chart title and axis labels
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('WorkWeekHrs')
plt.ylabel('Median Age')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
#lets have a quick look at Data
QUERY = """
SELECT ConvertedComp, Age 
FROM master
"""
#save it in dataframe 
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# Filter the data to include only ages between 45 and 60
filtered_data = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_comp = filtered_data.groupby('Age')['ConvertedComp'].median()
median_comp
```

```python
# Create a line plot of median ConvertedComp for each age
plt.plot(median_comp.index, median_comp.values)

# Set the x-axis label
plt.xlabel('Age')

# Set the y-axis label
plt.ylabel('Median ConvertedComp')

# Show the plot
plt.show()
```
